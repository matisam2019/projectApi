```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import pyautogui
import time
import pyperclip
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import json  # Para guardar el DOM como JSON

# Configura las opciones de Chrome (si necesitas usar un perfil específico)
options = Options()
# options.add_argument("--user-data-dir=/ruta/al/perfil/del/navegador")  # Descomenta y reemplaza si es necesario

# Inicializa el navegador (abre una nueva ventana)
driver = webdriver.Chrome(options=options)

# Navega a la URL deseada
driver.get("https://volvoconnect.com/login/")

# Maximiza la pantalla
driver.maximize_window()

# Espera a que la página cargue (puedes ajustar este tiempo)
time.sleep(5)  # Espera 10 segundos

# --- Código de pyautogui ---

# Tiempo de espera en segundos (puedes ajustarlo)
tiempo_espera_pyautogui = 5

# Función para escribir texto usando copy_paste
def escribir_texto(x, y, texto):
    pyautogui.moveTo(x, y)
    pyautogui.click()
    pyperclip.copy(texto)
    pyautogui.hotkey('ctrl', 'v')  # Windows
    # pyautogui.hotkey('command', 'v')  # macOS

# Función para simular la pulsación de Enter
def dar_enter(x, y):
    pyautogui.moveTo(x, y)
    pyautogui.click()
    pyautogui.press('enter')

# Espera inicial
print(f"Esperando {tiempo_espera_pyautogui} segundos...")
time.sleep(tiempo_espera_pyautogui)

print("Ejecutando acciones de escritura...")

# Escribir texto (ajusta las coordenadas según sea necesario)
escribir_texto(22, 567, 'sparra@jcabrera.cl')
escribir_texto(20, 648, "M@ti2711")
dar_enter(229, 712)

print("segundos para dar Enter...")

time.sleep(13)  # Espera 10 segundos
print("Simulando Enter...")
dar_enter(1137, 693)  # Selecciona Mapa
time.sleep(7) 
dar_enter(408, 409)
time.sleep(3) 
dar_enter(411, 464)





```

```python
try:
    # ... (Código de inicio de sesión como en el ejemplo anterior)

    # Función para guardar el DOM y eventos
    def guardar_dom_y_eventos(driver, iteracion):
        # Obtener el HTML del DOM
        html = driver.page_source

        # Guardar el HTML en un archivo (opcional)
        with open(f"dom_iteracion_{iteracion}.html", "w", encoding="utf-8") as f:
            f.write(html)

        # Convertir el DOM a JSON (opcional, para un análisis más estructurado)
        # Nota: Esto puede ser costoso en términos de rendimiento para DOMs grandes
        # dom_json = driver.execute_script("return document.documentElement.outerHTML;")
        # with open(f"dom_iteracion_{iteracion}.json", "w", encoding="utf-8") as f:
        #     json.dump(dom_json, f, indent=4)  # Formateado para mejor legibilidad

        # Capturar eventos principales (ejemplo: logs de la consola del navegador)
        # Puedes personalizar esto para capturar otros eventos relevantes
        logs = driver.get_log('browser')  # Requiere activar logs en ChromeOptions
        with open(f"eventos_iteracion_{iteracion}.json", "w") as f:
            json.dump(logs, f, indent=4)

        print(f"DOM y eventos de la iteración {iteracion} guardados.")

    # Iteraciones
    for i in range(1, 6):  # 4 iteraciones
        if i == 1:
            dar_enter(408, 409)
        elif i == 2:
            dar_enter(411, 464)
        elif i == 3:
            dar_enter(410, 565)
        elif i == 4:
            dar_enter(403, 643)
        elif i == 5:
            dar_enter(409, 729)
        time.sleep(6)
        guardar_dom_y_eventos(driver, i)

except Exception as e:
    print(f"Error: {e}")
```
