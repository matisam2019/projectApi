```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pyautogui
import pyperclip
import time
import keyboard
import csv
import pandas as pd
import sqlite3
from threading import Thread, Event

# Configure Chrome options
options = Options()

# Define DataFrame columns
columnas = [
    'id_ciclo', 'nombre', 'ubicacion', 'coordenadas', '% nivel combustible',
    'km hasta vacio', '% nivel de adblue', 'distancia(km)', 'horas motor(h)',
    'peso_camion(Ton)', 'peso_remolque(Ton)', 'actualizacion', 'conductor'
]

# Create empty DataFrame
df = pd.DataFrame(columns=columnas)

# Initialize database connection
conn = sqlite3.connect('datos_vehiculos.db')
cursor = conn.cursor()

# Initialize browser
driver = webdriver.Chrome(options=options)
driver.get("https://volvoconnect.com/login/")
driver.maximize_window()

def escribir_texto(x, y, texto):
    """Function to write text using copy-paste"""
    pyautogui.moveTo(x, y)
    pyautogui.click()
    pyperclip.copy(texto)
    pyautogui.hotkey('ctrl', 'v')

def observar_dom(stop_event):
    """Function to observe DOM changes until ESC is pressed"""
    elementos_cargados = []
    nombre_archivo_csv = 'elementos_cargados1.csv'
    
    with open(nombre_archivo_csv, 'w', newline='', encoding='utf-8') as archivo_csv:
        escritor_csv = csv.writer(archivo_csv)
        escritor_csv.writerow(['tag', 'attributes', 'text'])
        
        while not stop_event.is_set():
            try:
                # Get current DOM elements
                elements = driver.find_elements(By.XPATH, "//*")
                for element in elements:
                    try:
                        elemento_info = {
                            'tag': element.tag_name,
                            'attributes': element.get_attribute('outerHTML'),
                            'text': element.text.strip()
                        }
                        
                        # Solo guardar elementos que no estén ya en la lista
                        if elemento_info not in elementos_cargados:
                            elementos_cargados.append(elemento_info)
                            escritor_csv.writerow([
                                elemento_info['tag'],
                                elemento_info['attributes'],
                                elemento_info['text']
                            ])
                            print(f"Nuevo elemento encontrado: {elemento_info['tag']}")
                            
                    except Exception as e:
                        print(f"Error procesando elemento: {e}")
                        
                time.sleep(0.5)  # Pequeña pausa para no sobrecargar el CPU
                
            except Exception as e:
                print(f"Error observando DOM: {e}")
                
    return elementos_cargados

def dar_enter(x, y):
    """Enhanced enter function with DOM observation until ESC"""
    pyautogui.moveTo(x, y)
    pyautogui.click()
    pyautogui.press('enter')
    
    # Crear evento para detener la observación
    stop_event = Event()
    
    # Iniciar thread de observación
    observer_thread = Thread(target=observar_dom, args=(stop_event,))
    observer_thread.start()
    
    print("Observando cambios en el DOM. Presiona ESC para detener...")
    
    # Esperar hasta que se presione ESC
    while not keyboard.is_pressed('esc'):
        time.sleep(0.1)
    
    # Detener la observación
    stop_event.set()
    observer_thread.join()
    print("Observación del DOM detenida")

# Main execution
try:
    time.sleep(3)  # Initial wait for page load
    
    # Login sequence
    escribir_texto(22, 567, 'sparra@jcabrera.cl')
    escribir_texto(20, 648, "M@ti2711")
    dar_enter(229, 712)
    
except Exception as e:
    print(f"Error in main execution: {e}")
finally:
    # Cleanup
    conn.close()
    driver.quit()
```

```python
from selenium.webdriver.common.keys import Keys

def manejar_filtro_react(vehiculo_id):
    try:
        # Localizar el input
        input_element = WebDriverWait(driver, 15).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, 'input#positioningSearch'))
        )
        
        # Paso 1: Limpieza robusta
        input_element.click()
        input_element.send_keys(Keys.CONTROL + "a")  # Seleccionar todo el texto
        input_element.send_keys(Keys.DELETE)  # Eliminar selección
        
        # Disparar eventos React manualmente
        driver.execute_script("""
            const input = arguments[0];
            const nativeInputValueSetter = Object.getOwnPropertyDescriptor(
                HTMLInputElement.prototype, 
                'value'
            ).set;
            
            nativeInputValueSetter.call(input, '');
            input.dispatchEvent(new Event('input', { bubbles: true }));
            input.dispatchEvent(new Event('change', { bubbles: true }));
        """, input_element)
        
        # Paso 2: Escribir nuevo valor con eventos
        for char in vehiculo_id:
            input_element.send_keys(char)
            time.sleep(0.1)  # Simular escritura humana
            
        # Forzar actualización del estado React
        driver.execute_script("""
            arguments[0].dispatchEvent(new Event('blur'));
            arguments[0].dispatchEvent(new Event('focus'));
        """, input_element)
        
        # Verificación final
        WebDriverWait(driver, 10).until(
            lambda d: input_element.get_attribute('value') == vehiculo_id
        )
        
        print(f"✅ Filtro actualizado a: {vehiculo_id}")
        return True
        
    except Exception as e:
        print(f"🚨 Error crítico: {str(e)[:200]}")
        driver.save_screenshot(f'error_filtro_{vehiculo_id}.png')
        return False
```
